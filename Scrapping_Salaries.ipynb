DB salaries for 2021/2022

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
players = []
team = []
salary = []
lista = ['', 'page/2', 'page/3', 'page/4', 'page/5', 'page/6', 'page/7', 'page/8', 'page/9', 'page/10', 'page/11', 'page/12', 'page/13']

for j in lista:

    url = f'http://www.espn.com/nba/salaries/_/{j}/seasontype/3'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    td = soup.find_all('td')
    for i in list(td):
        if 'player' in str(i):
            temp = i.get_text()
            players.append(temp)
        elif 'team' in str(i):
            temp = i.get_text()
            team.append(temp)
        elif '$' in str(i):
            temp = i.get_text()
            salary.append(temp)
        
        
df = pd.DataFrame(list(zip(players, team, salary)),
               columns =['Player', 'Team', 'Annual Salary'])
```
